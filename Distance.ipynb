```python
import pandas as pd
from math import sin, cos, sqrt, atan2, radians
import numpy as np
rnd=pd.read_csv("Teachers_Random_20.csv",low_memory=False)
dm=pd.read_csv("Data_Model",low_memory=False)
```

```python
teacher=rnd.loc[:,['statevoterid','RegistrationAddressLatitude','RegistrationAddressLongitude']]
voters=dm.loc[:,['statevoterid','RegistrationAddressLatitude','RegistrationAddressLongitude']]
```

```python
voters.isnull().mean()
#teacher
```

```python
# for index, row in t.iterrows():
# #for i,j in zip(t["RegistrationAddressLatitude"],t["RegistrationAddressLongitude"]):
#     tech = (row["RegistrationAddressLatitude"], row["RegistrationAddressLongitude"])
#     for index, row in v.iterrows():
#         voter = (row["RegistrationAddressLatitude"], row["RegistrationAddressLongitude"])
#         distance=great_circle(tech, voter).miles
#         print(distance)
from geopy.distance import great_circle
tt=teacher
vv=voters
#taa = [[0] * m for i in range(n)]

taa=[]
for i in range(0,20):
    for j in range(0,1555513):
        taa.append(great_circle((tt['RegistrationAddressLatitude'][i],tt['RegistrationAddressLongitude'][i]),(vv['RegistrationAddressLatitude'][j],vv['RegistrationAddressLongitude'][j])).miles)

```

```python
zz=np.asarray(taa)
aaa=zz.reshape(-1,20,order='F')
```

```python
ww=pd.DataFrame(aaa)
xx=[]
for i in range(0,20):
    for j in range(0,1555513):
        if ww[i][j] < 0.5:
            xx.append(tt['statevoterid'][i])
        else:
            xx.append(np.nan)
            
```

```python
qwe=np.asarray(xx)
wee=qwe.reshape(-1,20,order='F')
weer=pd.DataFrame(wee)
weer.head(2)
```

```python
weer.isnull().mean()
```
